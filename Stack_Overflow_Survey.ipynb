```python
#Task 6
#A
values=df_sal_NonNa['FormalEducation'].value_counts()
labels="bachelor's degree","Master’s degree","Some college/university study without earning a degree","Secondary school","Associate degree","Other doctoral degree","Professional degree","Primary/elementary school","I never completed any formal education"
plt.pie(values,labels=labels,autopct="%.1f%%")
plt.title("Classification of Developers by their Formal Education")
plt.show()
```

```python
#B
values=df_sal_NonNa['FormalEducation'].value_counts()
labels="bachelor's degree","Master’s degree","Some college/university study without earning a degree","Secondary school","Associate degree","Other doctoral degree","Professional degree","Primary/elementary school","I never completed any formal education"
pie_explode=(0,0,0,0,0,0,0.1,0.1,0.1)
pie_color=['lime','red','blue','grey','yellow','silver','green','white','black']
plt.pie(values,labels=labels,autopct="%.1f%%",explode=pie_explode,colors=pie_color)
plt.title("Classification of Developers by their Formal Education")
plt.show()
```
